```python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf

import pandas as pd

```

```python
CSV_COLUMN_NAMES = ['SepalLength','SepalWidth','PetalLength','PetalWidth','species']
SPECIES = ['setosa','versicolor','virginica']
#this block is to create our constants
```

```python
#to get our dataset
train_path =tf.keras.utils.get_file('iris_training_csv','https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv')
#this will grab the file from the link and save the file as iris_training_csv
test_path =tf.keras.utils.get_file('iris_test_csv','https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv')

train = pd.read_csv(train_path,names=CSV_COLUMN_NAMES,header=0)
test = pd.read_csv(test_path,names=CSV_COLUMN_NAMES,header=0)
#here keras is used to grab our dataset and is read into pandas dataframe
```

```python
test.head()
```

```python
train_y =train.pop('species')
test_y =test.pop('species')
train.head()
```

```python
#input block here we create the input fuction
def input_fn(features,label,training=True,batch_size=256):
  #convert the inputs to  a dataset
  dataset = tf.data.Dataset.from_tensor_slices((dict(features),label))

  if training:
    dataset = dataset.shuffle(1000).repeat()
  return dataset.batch(batch_size)
```

```python
#feature columns, here we descrie how to use the input Function

my_feature_columns=[]
for items in train.keys():
  my_feature_columns.append(tf.feature_column.numeric_column(key=items))
print (my_feature_columns)

# building the model using DNN classifier(Deep Neural Network)
classifier = tf.estimator.DNNClassifier(
    feature_columns=my_feature_columns,
    hidden_units=[30,10],
    n_classes=3)
#estimator is a model that holds a alot of premade models which dnn is one of these models
classifier.train(
    input_fn= lambda:input_fn(train,train_y,training=True),
    steps=5000)
classifier.evaluate(input_fn= lambda:input_fn(test,test_y,training=False),steps=5000)
```

```python
eval_result=classifier.evaluate(input_fn= lambda:input_fn(test,test_y,training=False),steps=5000)
print('\ntest set accuracy:{accuracy:0.3f}\n'.format(**eval_result))
```

```python
#class prediction
def input_fn(features,batch_size=256):
  return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

features = ['SepalLength','SepalWidth','PetalLength','PetalWidth']
predict = {}
print('Please input numerics')

for feature in features:
  valid = True
  while valid:
    val = input(feature + ": ")
    if not val.isdigit(): valid=False
  predict[feature]= [float(val)]

prediction = classifier.predict(input_fn=lambda:input_fn(predict))
for pre_dict in prediction:
  class_id = pre_dict['class_ids'][0]
  probability = pre_dict['probabilities'][class_id]
  print ('prediction is "{}"({:.1f}%)'.format(SPECIES[class_id], 100 * probability))
```
